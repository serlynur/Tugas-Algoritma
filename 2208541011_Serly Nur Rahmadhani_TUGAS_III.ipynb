OPERASI MATEMATIKA PADA PYTHON

```python
#PENJUMLAHAN
print (0+9)
print (1.4+78+95.4+34) #mencoba penjumlahan dengan angka desimal
print (999+888+777+666+555+444+333+222+111) #penjumlahan dengan tiga digit angka dan lebih dari 4 variabel
print (88.9+(74+32)) 
```

```python
#PENGURANGAN
print (93-23-5-80) #mencoba operasi pengurangan yang menghasilkan output bilangan negatif
print (28-(-45)) #pengurangan dengan bilangan negatif
print (0.9-3.2-(5-2.3)) #pengurangan dengan angka desimal dan pengelompokkan bilangan dengan tanda kurung

print (33+(45-56)) #operasi campuran yang menggunakan penjumlahan dan pengurangan serta pengelompokkan operasi pada kurung, yang terjalan terlebih dahulu yaitu didalam kurung kemudian dikerjakan yang berada diluar kurung
print (33+(-45)-56) #operasi bilangan yang dimana akan menghasilkan output bilangan negatif dengan operasi campuran antara penjumlahan dan pengurangan
```

```python
#PERKALIAN
print (3*12) #operasi perkalian pada umumnya
print (9*123*0) #operasi perkalian yang melibatkan angka nol, mencoba apakah output yang dihasilkan sesuai
print (5*(-7)*6) #operasi perkalian dengan bilangan negatif, berhasil
print (5.7*6.2*(-5)*2) #operasi dengan bilangan desimal dan negatif yang menghasilkan hasil desimal yg lengkap tidak dengan dua digit terakhir atau tiga digit terakhir

print (2*(6-7)*0.2*(3.2+1.5)) #menggabungkan operasi pnjumlahan,pengurangan,dan perkalian dan menginput bilangan desimal tanpa float
```

```python
#PEMBAGIAN
print (7/3)
print (7/1.5) #bisa dibagi dengan bilangan desimal
print (7/2/3) #membagi tiga bilangan

print ((2+3*5)/2) #dalam pemrograman pengerjaan dalam tanda kurung yg dikerjakan terlebih dahulu tetap berlaku 
```

```python
#PEMBULATAN PEMBAGIAN
print (7//2) #dihitung berapa kali angka yg bisa dibagi, jadi sisa pembagian tidak dihitung
print (7//1.5) #kalau pembagi merupakan bilangan desimal maka hasil juga bilangan desimal, tapi jika hasilnya bilangan bulat maka ditambahin dengan ,0 dibelakang bilangan bulat tsb dan pada pengerjaan ini sisa pembagian tetap tidak termasuk
```

```python
#MODULO 
print (7%3) #modulo merupakan hasil dari sisa pembagian yg dikerjakan 
print (((9+4)*4)%3) 
print (9+4*4%3) #urutan pengerjaan yaitu perkalian,modulo,penjumlahan
```

```python
#PEMANGKATAN
print (3**3)
print ((5-7)*2**2) #dalam hal ini pemangkatan yg berada diluar kurung dikerjakan duluan kemudian di operasikan dengan yg berada didalam kurung
```

```python
#ARGUMEN OPERATOR
x = 5
x += 1
print (x) 
#artinya x += 1 sama seperti x = x + 1, yang dimana x harus dideskripsikan dulu 

x = -5
x -= 2
print (x)
#pada bilangan negatif juga berlaku hal tsb

x = -5
x //= 2  
print (x)
#outputnya -3 karena hasil dari -5//2 yaitu -2.5 
#sehingga hasil tsb dibulatkan kebawah makanya outputnya itu -3 

```

OPERATOR PERBANDINGAN DI PYTHON

```python
x = 3 > 2 
print (x)
#output yg dihasilkan benar

X = 3 > 2
print (x)
#hasilnya tdk eror tapi output dari ini mengikuti dari sebelumnya, 
#karena di deskripsinya itu menggunakan X kapital dan yg di print itu huruf x kecil,
#jadi hasil yg kluar itu sama kek sebelumnya
#kalau sebelumnya keluar true maka yg kdua jg true, hasilnya false ketika deskripsi di baris pertama itu menggunakan X besar

x = 3 < 2
print (x)

x = 3 >= 2
print (x)

x = 3 <= 2
print (x)

x = 3 == 3
print (x)
#karena 3 = 3 maka hasilnya true itu benar

x = 3 == 4
print (x)
#karena 3 dan 4 tidak sama sehingga outputnya false itu benar

x = 3 != 4
print (x)
# tanda seru disini artinya "tidak", jadi != berarti "tidak sama dengan", karena 3 dan 4 tidak sama sehingga outputnya true itu benar

x = 3 != 3
print (x)
#false karena 3 itu bilangan yang sama

```

OPERATOR LOGIKA DI PYTHON


```python
#operator logika menggunakan nilai kebenaran dari "and" yg dimana kedua variabelnya harus bernilai benar, jika salah satu variabel bernilai salah maka nilainya akan salah


A = 23
B = 32
#mendeskripsikan variabel yang ingin dicari terletak diawal dapat berupa kata atau variabel yg mewakili

print (A > 22 and B > 15)
#operator logika menggunakan nilai kebenaran dari "and" yg dimana kedua variabelnya harus bernilai benar, 
#jika salah satu variabel bernilai salah maka nilainya akan salah
#pada contoh ini kedua variabelnya bernilai benar sehingga outputnya yaitu "True"

print (A >= 23 or B < 15)
#operator logika dengan "or" yaitu salah satu variabelnya harus bernilai benar atau keduanya bernilai benar maka outputnya benar
#jika kedua variabel bernilai salah maka outputnya salah
#pada contoh ini variabel A bernilai benar karena 23>=23 itu bernilai benar, dan variabel B bernilai benar, sehingga outputnya "True" bernilai benar

print (A < 15 or B < 15)
#variabel A bernilai "salah" karena 23 < 15 itu salah 
#variabel B bernilai "salah" karena 32 < 15 itu bernilai salah
#sehingga "salah" or "salah", karena keduanya bernilai salah maka output yg dihasilkan yaitu bernilai salah "False"

print (A != 15 and B > 15)
#pada kasus ini variabel A bernilai benar karena A tidak sama dengan 15 tapi bernilai lebih kecil dari 23
#pada variabel B disini bernilai benar karena 32 lebih besar dari 15 itu benar
#sehingga "benar" dan "benar", maka nilai kebenarannya yaitu benar "True"
```

OPERATOR IF PADA PYTHON

```python
data = float (input ("Banyaknya kasus bullying di provinsi  X : "))
if (data >= 500) :
    print ("Kasus bullying yg terjadi di provinsi X masih tergolong cukup tinggi")
    print ("Semangat kepada pelopor dalam menurunkan angka kasus yg terjadi")
#disetiap setelah if dan persyaratan data harus ada tanda "titik dua"
#jika data yg dimasukkan tidak memenuhi syarat pertama maka lanjut ke syarat kedua

elif (data >= 300) :
    print ("Kasus bullying di provinsi X masih tergolong banyak")
    print ("Ayo kita ciptakan provinsi yang bebas dari bullying")
#disini data yg dimasukkan tidak memenuhi syarat kedua jadi lanjut ke syarat selanjutnya

elif (data >= 100) :
    print ("Di provinsi X kasus bullying termasuk di tingkatan yang rendah")
    print ("Tetap jaga keutuhan dan kedamaian para anak-anak maupun remaja agar kasus bullying tidak meningkat")
#karena data yg dimasukkan memenuhi syarat disini maka jalannya program terhenti disini dan hasil yg kluar sesuai yg berada di syarat ketiga

elif (data <= 50) :
    print ("Wow provinsi X merupakan lokasi yang memiliki kasus bullying terendah diantara 34 provinsi")
    print ("Jangan biarkan angka kasus ini meningkat, tetap berada di posisi terendah dalam kasus bullying")
#karena data tadi sudah memenuhi di syarat ketiga sehingga untuk syarat keempat tida terpakai karena sudah stop di syarat ketiga
```

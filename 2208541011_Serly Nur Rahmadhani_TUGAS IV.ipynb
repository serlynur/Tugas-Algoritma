**WHILE LOOP**

While loop akan selalu dieksekusi selama nilainya True dan akan berhenti
saat nilainya menjadi False

```python
#contoh while loop

a = 0
while 6 > a:
  print (a * " * ")
  a += 1
```

```python
#latihan penggunaan while loop ke-1
#menghitung pada faktorial

n = 5
#memasukkan nilai  yg ingin dicari disini contohnya saya masukkan angka 5

faktorial = 1
#nilai faktorial disini di deskripsikan sama dengan 1 atau dimulai dari 1

i = 1
#mendeskripsikan nilai dari i dimulai dari 1

while i <= n:
#disini yg akan dicari yaitu dengan syarat i kurang dari sama dengan n
    faktorial *= i
    i += 1
    #ini merupakan rumus dari faktorial yg akan kita cari

print("Faktorial dari", n, "adalah", faktorial)
```

```python
#latihan while loop ke-2
#menghitung bilangan ganjil dari 1 sampai 50

i = 1
#deskripsikan terlebih dahulu variabel yg akan dicari

count = 0
#disini di deskripsikan bahwa perhitungan dimulai dari 0

while i <= 50:
    if i % 2 != 0:
        count += 1
    i += 1

#nah disini karena kita ingin mencari bil ganjil dengan batas 50
#karena itu kita tuliskan bahwa i kurang dari sama dengan 50
#agar bil ganjil yg keluar tidak lebih dari 50
#kemudian kita masukkan rumus yg jika angka tsb terpenuhi bil ganjil maka angka tsb akan terhitung


print("Jumlah bilangan ganjil dari 1 hingga 50 adalah", count)
```

```python
#latihan while loop ke-3
#mencari nilai maksimum dari sejumlah bilangan

nilai = [3, 7, 1, 8, 4, 9]
#langkah pertama masukkan data yg akan dikerjakan

nilai_max = nilai[0]
i = 1
#mendeskripsikan awal dimulai nilai maximum terlebih dahulu

while i < len(nilai):
    if nilai[i] > nilai_max:
        nilai_max = nilai[i]
    i += 1
#kemudia masukkan rumus untuk mencari nilai maximum

print("Nilai maksimum adalah", nilai_max)
#kemudian print nilai maximum
```

**LIST** 

**List** dapat digunakan saat kita akan merepresentasikan objek di dalam Python
baik berupa string maupun berupa angka. Ini akan banyak digunakan pada
saat kita mengolah data pada Python. List dapat dipanggil urutannya, bahkan
isinya dapat dirubah

```python
#contoh penggunaan list pada python

bilangan = ['bulat', 'rasional', 'kompleks', 'real', 'desimal']
#mendeskripsikan variabel yg berbentuk list

print (bilangan)
#memanggil isi dari list variabel

print (bilangan[0])
#memanggil list dari variabel ke-1

print (bilangan[3])
#memanggil urutan list ke-4

print (bilangan [-1])
#memanggil urutan list paling terakhir
#jika mempunyai banyak data dan ingin memanggil data  yg terdekat dari terakhir
#gunakan tanda minus (-) yg dihitung dari akhir index

bilangan [4] = 'irasional'
print (bilangan)
#mengganti index ke 4 dengan kata baru

print (bilangan [0:5:2])
#memanggil data ke-1 dengan batas 5 yaitu index ke 4 dengan selisih 2 index

print (bilangan)
#memanggil kembali list dari variabel dengan data terakhir yg dimasukkan
#termasuk data yg telah diganti sebelumnya
```

```python
#latihan list ke-1
#mencari nilai modus

data = [1, 2, 3, 1, 2, 3, 1, 2, 3]
modus = data.modus(1)
#disini kita perlu tuliskan angka berapa yg akan kita cari
#angka yg dimasukkan akan dihitung berapa kali angka tsb muncul dalam data
#kurang lebih seperti modus tetapi kita masukkan angka yg ingin dicari

print(modus)
```

**METODA LIST**

Karena list merupakan objek, maka keberadaannya dapat dimanipulasi dengan
menambahkan atau menghilangkan isi nya.

```python
#contoh metoda list pada python

nama = ['made', 'eka', ' dwi']
#mendeskripsikan isi dari variabel
nama.append ('sari')
#append disini berguna untuk menambahkan suatu isi list dari variabel yg akan kita kerjakan
print (nama)
#memanggil variabel 

nama.insert (0, 'dewi')
#menambahkan isi variabel disertai letak urutan dari list tsb
print (nama)

nama.remove ('eka')
#remove berguna untuk menghapus isi list dari variabel agar tidak perlu memasukkan ulang data dari awal
print (nama)

print ('eka' in nama)
#contoh panggilan isi list yg mempunyai output false
#karena nama dari eka ini tidak ada dalam list tsb 

print ('dewi' in nama)
#contoh panggilan isi list yg mempunyai output true
#karena nama dari dewi ini ada dalam isi list yg di deskripsikan td

print (len(nama))
#len berguna untuk menghitung ada brp index dalam variabel terakhir yg setelah dikerjakan
#len disini mulai menghitung dari angka 1
```

```python
#latihan list ke-1

data = [29, 3, 12, 32, 9, 1, 99]
print (len(data))
#menghitung jumlah elemen dalam sebuah list

data.append (4.3)
print (data)
#menambahkan elemen ke dalam sebuah list

data.remove(32)
print (data)
#menghapus elemen dari sebuah list

data.reverse ()
print(data)
#membalik urutan elemen dalam sebuah list

data.sort()
print(data)
#mengurutkan elemen dalam sebuah list secara berurut dari terkecil ke terbesar

print (data[-1])
#memanggil elemen terakhir
```

**FOR LOOP**

For loop dapat digunakan untuk memanggil semua isi list. 

```python
#contoh penggunaan for loop

buah = ["manggis", "durian", "salak", "jeruk"]
for jenis in buah:
#jenis disini deskripsi bebas yg sebagai pelengkap saja dalam memanggil elemen dari variabel
	print (jenis)
```

```python
#contoh penggunaa for loop

buah = ["manggis", "durian", "salak", "jeruk"]
for i in range (len(buah)):
#range disini sama penggunaannya seperti jenis diatas 
#memanggil sekaligus menghitung jumlah keseluruhan elemen dari variabel
	print (buah[i])
```

```python
#latihan for loop ke-1
#menampilkan angka dari 1 hingga 10

for i in range(1, 11):
#disini dimasukkan angka 11 karena kita ingin menampilkan angka hingga 10
#karena batas dari sbuah elemen dihitung dari 0
    print(i)
```

```python
#latihan for loop ke-2

data = [11, 22, 33, 44, 55, 66]
#mendekripsikan data terlebih dahulu 
jumlah = 0
#mendeskripsikan bahwa jumlah dimulai dari 0
total_elemen = 0
#mendeskripsikan bahwa total elemen dimulai dari 0

for angka in data:
    print(angka)
#menampilkan elemen dari sebuah list yg telah di deskripsikan sebelumnya


for angka in data:
    total_elemen += 1
print (total_elemen)
#menghitung jumlah elemen dari sebuah list
#karena dari deskripsi bahwa data yg kita punya 6 maka outputnya akan 6
#jika kita mempunyai banyak data nantinya itu akan berguna untuk menghitung jumlah data tse tanpa menghitung manual

for angka in data:
    jumlah += angka
print (jumlah)
#menjumlahkan elemen dalam sebuah list
#disini menghitung elemen tanpa menuliskan kembali elemen dari variabel tsb
```

**FUNGSI RANGE**

Fungsi range berguna untuk membangkitkan bilangan berurutan atau berpola
yang kita inginkan. Untuk menuliskannya, perlu dikombinasikan dengan for
loop. L

```python
#contoh fungsi range

angka = range (6)
print (angka)
#deskripsikan terlebih dahulu dengan range 0 sampai 6

print (" ")
#mencetak baris kosong atau enter

for isi in angka :
    print (isi)
#menampilkan elemen yg berada di range yg sebelumnya di deskripsikan yaitu 0 hingga 6

print (" ")
#mencetak enter

for isi in range (5,10):
    print (isi)
#menampilkan elemen dari 5 hingga 10

print (" ")
#mencetak enter

for isi in range (0,10,2):
    print (isi)
#menampilakn elemen dari 0 hingga 10 dengan selisih atau kelipatan 2 yg dimulai daari 0
```

```python
#latihan fungsi range ke-1
#menemukan jumlah bilangan prima antara dua angka

start = 2
end = 100
count = 0
#disini kita akan mencari bilangan prima antara dua angka dengan batas  sampai 100
#dengan itu dimulai dari angka 2 dan diakhiri dengan angka 100
#count disini berguna sebagai perhitungan yg dimulai dari 0

for i in range(start, end+1):
    if i > 1:
        for j in range(2, i):
            if (i % j) == 0:
                break
        else:
            count += 1
#baris pertama for disini dimulai dengan deskripsikan batas awal dan batas akhir
#kemudian buat variabel dsni saya menggunakan variabel i sebagai bilangan prima yg dicari
#selanjutnya saya membuat j didalam for didalam for untuk menghitung bilangan prima dengan selisih 2 angka dan range dari 2 karena kita akan mencari bil prima antara dua angka
#kemudian angka akan terhitung bilangan prima dari 2 dengan modulo 2 karena selisih 2
#kemudian count += 1 ini akan melanjutkan angka angka selanjutnya yg merupakan bilangan prima antara 2 angka

print(count)
#program akan berjalan
```

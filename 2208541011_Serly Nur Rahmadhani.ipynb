```python
THR = int (input ('Pendapatan THR : '))
gaji = int (input ('Pendapatan gaji sebulan : '))
makan = 2000000
kebutuhan = 1000000
#total = int (THR + gaji) #kalau menggunakan ini operasi penjumlahan tidak bisa
print ('Total uang saya : ' + str(THR + gaji)) #menggunakan str bila memanggil perhitungan
```

```python
print ('Total gaji selama setahun : ' + str (gaji * 12) )
print ('Total uang selama 10 tahun tanpa pengeluaran : ' + str (gaji * 12 + THR))
```

```python
total = int (makan) + int (kebutuhan) #penjumlahan int harus satu satu jika digabung tidak terhitung
print (str (total)) #harus masukkan rumus total dulu kemudian gunakan deskripsi total
```

```python
print ('Total uang selama setahun : ' + str ((gaji * 12) - total)) #tidak bisa digabung
```
